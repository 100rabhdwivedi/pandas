## 1. Converting Columns to Datetime (`pd.to_datetime`)
- Use `pd.to_datetime()` to convert strings or numbers into datetime objects.
- Once converted, Pandas recognizes them as dates, enabling powerful operations.

```python
import pandas as pd

data = {
    "OrderID": [101, 102, 103, 104, 105],
    "OrderDate": [
        "01-03-2026",
        "02/03/2026",
        "03-03-2026",
        "04-03-2026",
        "05/03/2026"
    ],
    "Amount": [2500, 1800, 3200, 1500, 2700],
    "Customer": ["Saurabh", "Amit", "Neha", "Raj", "Priya"]
}

df = pd.DataFrame(data)

df['OrderDate']=pd.to_datetime(df['OrderDate'] ,errors="coerce")
```

## 2. Setting DateTime Index & Resampling
- Set a datetime column as the index with `set_index()`.
- Resample time series using `resample('M').mean()` or other rules ('D', 'W', 'Y').
- Useful for aggregating data by time periods.

```python
df_idx = df.set_index('OrderDate')
df_idx.resample("Y")['Amount'].sum()
```

## mini-project 

Converting columns to datetime (pd.to_datetime)

Setting a DateTime index & resampling

```python
import pandas as pd

data = {
    "Date": [
        "2026-03-01",
        "2026-03-02",
        "2026-03-03",
        "2026-03-04",
        "2026-03-05",
        "2026-03-06",
        "2026-03-07"
    ],
    "Visit_Count": [120, 150, 180, 130, 200, 220, 170]
}

df = pd.DataFrame(data)

# 2️⃣ Convert Date to datetime
df["Date"] = pd.to_datetime(df["Date"])

# 3️⃣ Set Date as index
df = df.set_index("Date")

# 4️⃣ Monthly total visitors
monthly = df.resample("M")["Visit_Count"].sum()

# 5️⃣ Weekly average visitors
weekly = df.resample("W")["Visit_Count"].mean()

print("Monthly:\n", monthly)
print("\nWeekly:\n", weekly)
```
