## 1. Applying Custom Functions with `apply()` / `map()`
- `apply()` lets you apply a custom function across rows or columns.
- `map()` is for applying functions element-wise to a Series.
- Both are handy for transformations, calculations, and cleaning data.

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Saurabh", "Amit", "Neha", "Raj", "Priya"],
    "Science": [85, 78, 92, 74, 88],
    "Math": [90, 82, 89, 70, 25]
}

df = pd.DataFrame(data)

df['Total Marks'] = df[['Science','Math']].apply(lambda row : row.sum(),axis=1)
df
```

```python
col_mean = df[['Science','Math']].apply(lambda col : col.mean())
col_mean

def to_grade(s):
    if s>90:
        return "A"
    elif s>70:
        return "B"
    elif s>40:
        return "C"
    else:
        return "F"
    
df['Math Grade'] = df['Math'].apply(to_grade)
df['Science Grade'] = df['Science'].apply(to_grade)
df['Name'] = df['Name'].map(str.lower)
df
```
