Welcome to **Class 4** of our Pandas series. Today we’ll explore techniques to select and filter data in a DataFrame.

## 1. Boolean Indexing & Conditional Selection
- Use comparison operators to build conditions (e.g., `df['col'] > 50`).
- Pass the condition inside `df[...]` to get matching rows.
- Combine multiple conditions with `&` (AND), `|` (OR), and `~` (NOT).
- Parentheses are required when combining conditions.

```python

import pandas as pd
# 1. Build a DataFrame with columns: Name, Age, City, Score

data = {
    "Name":['saurabh','gaurav','pavan','vikash','vijay'],
    "Age":[20,25,30,28,32],
    "City":['Pune','Delhi','Rewa','Mumbai','Banglore'],
    "Score":[90,95,80,60,30]
}
df = pd.DataFrame(data)
df['Age'] > 25
df[df['Score'] == 95]

df[(df['Age']>25) | (df['Score'] != 40) & (df['City'] != 'Pune' ) ]
df[~(df['Age']>25)]
```

## 2. Filtering Rows with Multiple Conditions
- Combine two or more conditions to refine selection.
- Remember to enclose each condition in parentheses.

## 3. Using `query()` for Cleaner Filtering
- `query()` allows filtering rows with a string expression.
- It's concise and readable for complex conditions.
- Variables can be passed using the `@` symbol.

```python
df.query('Score >=85')
```

## Mini Practice
1. Build a DataFrame with columns such as Name, Age, City, Score.
2. Use boolean indexing to select rows where Age > 25.
3. Filter rows where City is either 'Delhi' or 'Mumbai'.
4. Use query() to find rows with Score >= 85 and Age < 35.
5. Clean a text column using str.replace or str.lower.

```python


import pandas as pd
# 1. Build a DataFrame with columns: Name, Age, City, Score

data = {
    "Name":['saurabh','gaurav','pavan','vikash','vijay'],
    "Age":[20,25,30,28,32],
    "City":['Pune','Delhi','Rewa','Mumbai','Banglore'],
    "Score":[90,95,80,60,30]
}
df = pd.DataFrame(data)
# 2. Use boolean indexing to select rows where Age > 25
# Write your code here

print(df[df['Age']>25])

# 3. Filter rows where City is either 'Delhi' or 'Mumbai'
# Write your code here

print(df[(df['City']=='Mumbai')|(df['City'] == 'Delhi')])

# 4. Use query() to find rows with Score >= 85 and Age < 35
# Write your code here

print(df.query("Score >= 85 and Age <35"))

# 5. Clean a text column using str.replace or str.lower
# Example: convert City names to lowercase or replace "delhi" with "New Delhi"
# Write your code here
df['City'] = df['City'].str.lower()
print(df['City'].str.replace('delhi','New Delhi'))
```
